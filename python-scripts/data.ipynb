```python
df[['calories','total fat (PDV)','sugar (PDV)','sodium (PDV)','protein (PDV)','saturated fat (PDV)','carbohydrates (PDV)']] = df.nutrition.str.split(",",expand=True) 
df['calories'] =  df['calories'].apply(lambda x: x.replace('[','')) 
df['carbohydrates (PDV)'] =  df['carbohydrates (PDV)'].apply(lambda x: x.replace(']','')) 
df[['calories','total fat (PDV)','sugar (PDV)','sodium (PDV)','protein (PDV)','saturated fat (PDV)','carbohydrates (PDV)']] = df[['calories','total fat (PDV)','sugar (PDV)','sodium (PDV)','protein (PDV)','saturated fat (PDV)','carbohydrates (PDV)']].astype('float')
df.head()
```

Creation de la matrice user/recette = notes

```python
# ouverture du tableau de note des utilisateurs
# besoin du fichier "RAW_interactions.csv"

df = pd.read_csv("RAW_interactions.csv")
grade = df[["user_id","recipe_id","rating"]]

#creation d'un dataframe permettant de choisir les utilisateurs avec plus de 50 votes

data_df = pd.DataFrame(grade['user_id'].value_counts())
final_data = data_df[(data_df["user_id"]>50)]
index_list = final_data.index
in_index = grade[(grade["user_id"].isin(index_list))]

#création de la matrice user/recette = notes
matrice = in_index.pivot(index='recipe_id', columns='user_id', values='rating')
matrice
```
